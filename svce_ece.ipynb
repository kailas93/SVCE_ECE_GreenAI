```python
# sum of digits in a given number using while loop 
# input ---> 1234
# output ---> 10 
num = int(input("enter a number: "))
s = 0
while num>0:
    digit = num%10
    s = s+ digit
    num = num//10
print(s)    
```

```python
#product of digits in a given number using while loop
# input ---> 1234
# output ---> 24
num = int(input("enter a number: "))
prod = 1
while num>0:
    digit = num%10
    prod = prod* digit
    num = num//10
print(prod) 
```

```python
# find the sum of digits in a given number using for loop
n = input("enter a number:  ")
import time
start = time.perf_counter()
s = 0
for i in n:
    s = s+int(i)
print(s)
end = time.perf_counter()
print("total time: ",end - start)
```
